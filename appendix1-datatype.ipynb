# **데이터 타입 조정**



*   변수(칼럼) 타입 확인 : 데이터.dtypes
*   변수를 범주형으로 변경 :
  - 데이터['칼럼명'].astype('category')
*   변수를 수치형으로 변경 :
  - 데이터['칼럼명'].astype('int')
  - 데이터['칼럼명'].astype('float')
*   NA 값의 처리
  - NA 갯수 체크 : 데이터.isna().sum()
  - NA 값 채우기 : 데이터['칼럼명'].fillna(특정숫자)

```
import pandas as pd
```

```
file_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
iris = pd.read_csv(file_url)
iris.head()
```

```
# 원핫인코딩 => (안됨)품종 column에 0, 1, 2의 숫자가 들어있기 때문에 범주형 데이터로 인지하지 못함
encoding = pd.get_dummies(iris)
encoding.head()
```

```
print(iris.dtypes)
```

```
# 품종 타입을 범주형으로 바꾸어준다.
iris['품종'] = iris['품종'].astype('category')
print(iris.dtypes)
```

```
# 다시 원핫인코딩
encoding = pd.get_dummies(iris)
encoding.head()
```

```
# NA값 체크
iris.isna().sum()
```

```
# NA값이 있으면 model이 읽어들일 수 없어 오류가 남
# => 값을 바꿔주거나 행을 비워야 한다.
iris.tail()
```

```
# NA값에 꽃잎폭 평균값을 넣어주는 방법
mean = iris['꽃잎폭'].mean()
print(mean)
iris['꽃잎폭'] = iris['꽃잎폭'].fillna(mean)
iris.tail()
```
