레몬에이드 판매 예측



*   git csv url : 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'





```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터 준비
file_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
data = pd.read_csv(file_url)
data.head()
```

```
# 종속변수, 독립변수 분리
ind_var = data[['온도']]
d_var = data[['판매량']]
print(ind_var.shape, d_var.shape)
```

```
# 모델 만들기
X = tf.keras.layers.Input(shape=[1]) # ind_var의 column의 수가 1
Y = tf.keras.layers.Dense(1)(X) # d_var의 column의 수가 1
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# 모델 학습
# verbose=0 -> 학습하는 동안 화면 출력X
model.fit(ind_var, d_var, epochs=10000, verbose=0) 
```

```
# 모델 이용
model.predict(ind_var)
```

```
# 종속변수(온도)가 15일 경우 몇 개의 레몬에이드를 준비해야하는지 예측
model.predict([[15]])
```
