보스턴 집값 예측

```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 1. 과거의 데이터 준비
file_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(file_url)
print(boston.columns)
boston.head()
```

```
# 독립변수, 종속변수 분리
id_var = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
d_var = boston[['medv']]
print(id_var.shape, d_var.shape)
```

```
# 2. 모델의 구조 만들기
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8, activation="swish")(X)
H = tf.keras.layers.Dense(8, activation="swish")(X)
H = tf.keras.layers.Dense(8, activation="swish")(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# Activation을 분리해서 적어준다.(위의 모델과 동일)
# Batchn normalization을 Dense layer와 Activation layer 사이에 적어주는 것이 좋다.

X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# 4. 데이터로 모델 학습(FIT)
model.fit(id_var, d_var, epochs=1000)
```

```
model.summary()
```

```
# 5. 모델 이용하기
print(model.predict(id_var[:5]))
print(d_var[:5])
```

```
d_var[0:5]
```

```
# 모델의 수식 확인
model.get_weights()
```

아이리스 품종 분류

```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 1. 과거의 데이터 준비
file_url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(file_url)
iris.head()
```

```
# 원핫인코딩
encoding = pd.get_dummies(iris)
encoding.head()
```

```
print(encoding.columns)
```

```
# 독립변수, 종속변수 분리
id_var = encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
d_var = encoding[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(id_var.shape, d_var.shape)
```

```
# 2. 모델 구조 만들기
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(X)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')
```

```
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')
```

```
# 3. 모델 학습(FIT)
model.fit(id_var, d_var, epochs=1000)
```

```
# 4. 모델 이용 
print(model.predict(id_var[0:5]))
print(d_var[0:5])
```

```
print(d_var[-5:])
```

```
# 가중치 출력
model.get_weights()
```
